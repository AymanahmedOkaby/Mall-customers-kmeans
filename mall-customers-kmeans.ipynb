# Albow method to find the number of clusters

```python

sse = []
k_rng = range(1,10)
for k in k_rng:
    kn = KMeans(n_clusters=k)
    kn.fit(df[['Spending Score (1-100)','Annual Income (k$)']])
    sse.append(kn.inertia_)
#plt.xlabel('K')
#plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```

# the model

```python
kn = KMeans(n_clusters=5)
y_predicte=kn.fit_predict(df[['Annual Income (k$)','Spending Score (1-100)']])
y_predicte
```

```python
df['cluster']=y_predicte
```
